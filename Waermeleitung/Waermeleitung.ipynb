# Wärmeleitung und thermoelektrische Effekte

## Motivation

Als [Wärmeleitung](https://de.wikipedia.org/wiki/W%C3%A4rmeleitung) (Konduktion) versteht man den Transport von Wärme (ohne Stoffaustausch), in oder zwischen Stoffen, aufgrund von Temperaturunterschieden. In unserer Modellvorstellung erfolgt dieser Austausch durch (Gitter-)Schwingungen oder Stöße einzelner Partikel aus denen ein Stoff besteht. Andere Formen des Wärmeaustauschs bestehen in der [Konvektion](https://de.wikipedia.org/wiki/Konvektion) (mit Stoffaustausch) und der Emission/Absorption von [Wärmestrahlung](https://de.wikipedia.org/wiki/W%C3%A4rmestrahlung). Wärmeleitung erfolgt immer in Richtung der niedrigeren Temperatur. Verschiedene Stoffe weisen unterschiedlich gute Wärmeleitungseigenschaften auf, die man durch ihre [Wärmeleitfähigkeit](https://de.wikipedia.org/wiki/W%C3%A4rmeleitf%C3%A4higkeit) als Materialeigenschaft quantifiziert. Die Wärmeleitfähigkeit und Wärmeleitung von Stoffen zu bestimmen ist ein sehr technisches Feld. Eine quantitative Beschreibung erfolgt mit Hilfe des [Fourierschen Gesetzes](https://de.wikipedia.org/wiki/W%C3%A4rmeleitung#W%C3%A4rmestrom,_Fouriersches_Gesetz) oder der [Wärmeleitungsgleichung](https://de.wikipedia.org/wiki/W%C3%A4rmeleitungsgleichung), deren Lösung i.a. schon bei einfachsten Stoffen und Oberflächen nur mit Hilfe numerischer Methoden möglich ist. In unserem Alltag spielt Wärmeleitung v.a. bei der Wärmedämmung oder Isolation, z.B. beim Hausbau oder bei Kühlaggregaten eine wichtige Rolle. In der Forschung geht es oft darum die unerwünschte Abwärme z.B. eines Detektors, einer tiefgekühlten Messanordnung oder eines Auslesechips abzuleiten. Dies erfolgt, in einfachsten Fällen, durch einen Kaltwasserstrom; bei elektrischen Bauteilen mit Hilfe von Luft oder Kühlgasen (wie z.B. [Trockeneis](https://de.wikipedia.org/wiki/Trockeneis)). Schwer zugängliche Bauteile können mit Stoffen mit hoher Wärmeleitfähigkeit eng verbunden werden, die die entstehende Wärme abführen und an anderer Stelle gekühlt werden. 

Durch das Verrichten von Arbeit kann Wärme auch bewusst transportiert werden. Ein solches Vorgehen kann auch zur *aktiven* Kühlung verwendet werden. Im Laufe dieses Versuchs werden Sie den [Peletier-Effekt](https://de.wikipedia.org/wiki/Peltier-Element), als einen [thermoelektrischen Effekt](https://de.wikipedia.org/wiki/Thermoelektrizit%C3%A4t) studieren können, mit dessen Hilfe man Wärme durch elektrische Arbeit transportieren, oder umgekehrt durch Wärmetransport (beeindruckend große) elektrische Arbeit verrichten kann.

```python
import numpy as np #Lade Pakete
import matplotlib.pyplot as plt
from PhyPraKit import readPicoScope, resample
from scipy import signal
from scipy import interpolate
import sys
from kafe2 import  XYContainer, Fit, XYFit, Plot, ContoursProfiler                            
from uncertainties import ufloat, unumpy
import uncertainties.unumpy as unp
from math import log10, floor
import scipy.integrate as integrate
import pandas as pd
import csv
```

```python
r_stab_Cu = ufloat (0.008,0.001 ) # Radius Kupferstab in m
abstand_mess_Cu = ufloat (0.02 ,0.001 ) # Abstände der Messfühler Kupfer in m
T_kühlwasser_Cu = ufloat ( 17.8,0.1 ) #Temperatur Kühlwasser Kupfer in °C
I_heizspule_Cu = ufloat ( 0.58,0.005 )  #Strom Heizspule Kupfer in A
U_heizspule_Cu = ufloat ( 7.8,0.05 )  #Spannung Heizspule Kupfer in V
strecke_mess_Cu = unumpy.uarray([ 0.,0.02 ,0.04 ,0.06 ,0.08 ],np.full(5, 0.001)) #Steckenmessung Kupfer in m, Beginn bei T_min
T_mess_Cu = unumpy.uarray([42.07,42.97,44.06,44.99,46.03],np.full(5, 0.02 )) # Temperaturmessung Kupfer in °C

r_stab_Stahl = ufloat (0.008, 0.001) # Radius Stahlstab in m
abstand_mess_Stahl = ufloat ( 0.02,0.001 ) # Abstände der Messfühler Stahl in m
T_kühlwasser_Stahl = ufloat ( 17.2,0.1 ) #Temperatur Kühlwasser Stahl in °C
I_heizspule_Stahl = ufloat ( 0.59,0.005 )  #Strom Heizspule Stahl in A
U_heizspule_Stahl = ufloat ( 8.0,0.05 )  #Spannung Heizspule Stahl in V
strecke_mess_Stahl = unumpy.uarray([0 ,0.02 ,0.04 ,0.06 ,0.08 ],np.full(5,0.001 )) #Steckenmessung Stahl in m, Beginn bei T_min
T_mess_Stahl = unumpy.uarray([42.63 ,48.52 ,54.60 ,58.20 , 69.49],np.full(5,0.02 )) # Temperaturmessung Stahl in °C

r_stab_Messing = ufloat (0.008, 0.001) # Messing Kupferstab in m
abstand_mess_Messing = ufloat (0.02 ,0.001 ) # Abstände der Messfühler Messing in m
T_kühlwasser_Messing = ufloat ( 17.2,0.1 ) #Temperatur Kühlwasser Messing in °C
I_heizspule_Messing = ufloat (0.59 ,0.005 )  #Strom Heizspule Messing in A
U_heizspule_Messing = ufloat (8.0 , 0.05 )  #Spannung Heizspule Messing in V
strecke_mess_Messing = unumpy.uarray([ 0,0.02 ,0.04 ,0.06 ,0.08 ],np.full(5,0.001 )) #Steckenmessung Messing in m, Beginn bei T_min
T_mess_Messing = unumpy.uarray([40.85 ,45.58 ,51.84 ,51.86 ,58.3 ],np.full(5,0.001 )) # Temperaturmessung Messing in °C
```

Um aus den gemssenen Daten die Wärmeleitfahigkeit der Materialen $\lambda$ zu bestimmen, wird die folgende Formel (https://de.wikipedia.org/wiki/W%C3%A4rmeleitf%C3%A4higkeit) benötigt:
$$\lambda = \frac{P\cdot l}{A \cdot \Delta T} = \frac{U_{heizspule} \cdot I_{heizspule}\cdot strecke_{mess}}{\pi \cdot r_{stab}^2 \cdot \left(T_{mess}-T_{kühlwasser}\right)}$$  \
Die Unsicherheiten werden mit Hilfe Gaußscher Fehlerfortpflanzung berechnet. Dies geschieht mit dem Python-Paket "uncertainties". Das Paket hat die Fehlerfortpflanzung schon implementiert und rechnet die Unsicherheiten automatisch aus. 

```python
#kupfer
vorfaktor = U_heizspule_Cu*I_heizspule_Cu/(np.pi*r_stab_Cu**2)
vorfaktor = unumpy.nominal_values(vorfaktor)

xy_data = XYContainer(unumpy.nominal_values(strecke_mess_Cu),unumpy.nominal_values(T_mess_Cu)-unumpy.nominal_values(T_kühlwasser_Cu))

def linear(x,a=10,b=10):
    return vorfaktor/a*x + b
line_fit=Fit(data=xy_data,model_function=linear)
line_fit.add_error(axis='x', err_val = unumpy.std_devs(strecke_mess_Cu))
line_fit.add_error(axis='y', err_val = unumpy.std_devs(T_mess_Cu-T_kühlwasser_Cu))
line_fit.do_fit()
line_fit.assign_parameter_latex_names(x='strecke', a='\lambda', b ='b')
line_fit.assign_model_function_latex_name('')
line_fit.assign_model_function_latex_expression('{a} \\cdot {x} + {b}')


line_fit.data_container.label = "Kupfer"
line_fit.model_label = "Modellfunktion"

line_fit.data_container.axis_labels = [ "Strecke in m","T_mess-T_kühl in °C"]
plot = Plot(fit_objects =line_fit)
plot.plot()
plt.show()

#messing
vorfaktor = U_heizspule_Messing*I_heizspule_Messing/(np.pi*r_stab_Messing**2)
vorfaktor = unumpy.nominal_values(vorfaktor)

xy_data = XYContainer(unumpy.nominal_values(strecke_mess_Messing),unumpy.nominal_values(T_mess_Messing)-unumpy.nominal_values(T_kühlwasser_Messing))

def linear(x,a=10,b=10):
    return vorfaktor/a*x + b
line_fit=Fit(data=xy_data,model_function=linear)
line_fit.add_error(axis='x', err_val = unumpy.std_devs(strecke_mess_Messing))
line_fit.add_error(axis='y', err_val = unumpy.std_devs(T_mess_Messing-T_kühlwasser_Messing))
line_fit.do_fit()
line_fit.assign_parameter_latex_names(x='strecke', a='\lambda', b ='b')
line_fit.assign_model_function_latex_name('')
line_fit.assign_model_function_latex_expression('{a} \\cdot {x} + {b}')


line_fit.data_container.label = "Messing"
line_fit.model_label = "Modellfunktion"

line_fit.data_container.axis_labels = [ "Strecke in m","T_mess-T_kühl in °C"]
plot = Plot(fit_objects =line_fit)
plot.plot()
plt.show()


#stahl
vorfaktor = U_heizspule_Stahl*I_heizspule_Stahl/(np.pi*r_stab_Stahl**2)
vorfaktor = unumpy.nominal_values(vorfaktor)

xy_data = XYContainer(unumpy.nominal_values(strecke_mess_Stahl),unumpy.nominal_values(T_mess_Stahl)-unumpy.nominal_values(T_kühlwasser_Stahl))

def linear(x,a=10,b=10):
    return vorfaktor/a*x + b
line_fit=Fit(data=xy_data,model_function=linear)
line_fit.add_error(axis='x', err_val = unumpy.std_devs(strecke_mess_Stahl))
line_fit.add_error(axis='y', err_val = unumpy.std_devs(T_mess_Stahl-T_kühlwasser_Stahl))
line_fit.do_fit()
line_fit.assign_parameter_latex_names(x='strecke', a='\lambda', b ='b')
line_fit.assign_model_function_latex_name('')
line_fit.assign_model_function_latex_expression('{a} \\cdot {x} + {b}')


line_fit.data_container.label = "Stahl"
line_fit.model_label = "Modellfunktion"

line_fit.data_container.axis_labels = [ "Strecke in m","T_mess-T_kühl in °C"]
plot = Plot(fit_objects =line_fit)
plot.plot()
plt.show()



```

**Lösung:**

In dieser Aufgabe wurde ein nach außen hin isolierter Peletier-Block, an den eine variable Spannung angelegt werden kann, betrieben. Dieser kühlt die eine Seite des Blocks herunter. Die andere Seite wird mit Wasser auf konstanter Temperatur gehalten. Dann wird die Temperaturdifferenz der beiden Bauelemente bestimmt.

```python
I_block = unumpy.uarray ([0,5.0, 10, 15.1, 20.0],np.full(5,0.1 )) #gemessener Strom am Block in A
U_block = unumpy.uarray ([0,0.5, 1.1,1.7,2.3  ],np.full(5, 0.1)) #gemessene Spannung am Block in V
T_1 = unumpy.uarray ([38.88,22.01,3.07, -0.5, 1.34],np.full(5,0.01 )) # Temperatur für Spannungsdifferenz in V
T_2 = unumpy.uarray ([40.95,36.24,32.32, 34.28,39.66],np.full(5, 0.01)) # Temperatur für Spannungsdifferenz in V, T_2>T_1
delta_T = T_2-T_1
print(delta_T)
#delta T über I_block fitten nur für verlauf exp



xy_data = XYContainer(unumpy.nominal_values(I_block),unumpy.nominal_values(delta_T))
def exp(x,a=-45.,b=44.,c=-0.102):
    return a*np.exp(c*x)+b
line_fit=Fit(data=xy_data,model_function=exp)

line_fit.add_error(axis='x', err_val = 0.5)
line_fit.add_error(axis='y', err_val = 0.5)

line_fit.do_fit()

#line_fit.assign_parameter_latex_names(x='delta T in °C', a='a',b='b')
#line_fit.assign_model_function_latex_name('')
#line_fit.assign_model_function_latex_expression('{a} \\cdot {x}')


line_fit.data_container.label = "Verwendete Daten"
line_fit.model_label = "Modellfunktion"

line_fit.data_container.axis_labels = [ "I_Kühl in A","delta T in °C"]
plot = Plot(fit_objects = line_fit)  
plot.plot()
plt.show()

```

Mit einem exponentiellen Fit ist die Temperaturdifferenz in Abhängigkeit der Stromstärke des Petierelements sichtbar. Die Abhängigkeit folgt der Funktion $\Delta T \left (I\right)=-45.0^{\circ}C\cdot e^{-0.102\frac{1}{A}\cdot I}+44.4^{\circ}C$. Damit folgt, dass das Petierelement eine maximale Temperaturdifferenz von ungefähr $44,4\pm 2,5^{\circ}C$ erzeugen kann.
Die beschränkte Temperaturdifferenz kann an der Raumtemperatur liegen, die das Petierelement trotz Isolierung nach außen aufwärmt. Außerdem führt der elektrische Widerstand des Petierelements zu Wärmeerzeugung, die die Temperaturdifferenz hemmt.

Führen Sie auf jeden Fall eine Messung bei $\Delta T=3^{\circ}\mathrm{C}$ durch. Wenn es die Zeit erlaubt, ist eine Vergleichsmessung bei $\Delta T=6^{\circ}\mathrm{C}$ ebenfalls interessant. 

---

**Lösung:** \
Im zweiten Versuchsteil wird an an einer Seite des Blocks ein Heizelement angeschalten. Dann wird an dem Heizelement die Spannung so lange erhöht/erniedrigt bis eine Temperaturdifferenz von $3^{\circ}C$ sich zwischen den Blockenden einstellt. Bei dieser Temperaturdifferenz wird der Kühlstrom, die Kühlspannung, der Heizstrom und die Heizspannung gemessen.

```python
I_kühl = unumpy.uarray([5, 10, 15.1,20.0],np.full(4, 0.5)) #Kühlstrom in A
U_kühl = unumpy.uarray([0.5, 1, 1.6,2.2],np.full(4,0.1 )) #Kühlspannung in V
I_heiz = unumpy.uarray([1.23,1.6,2.9,2.18],np.full(4,0.5 ))   #Heizstrom in A
U_heiz = unumpy.uarray([5.9,6.4,6.8,9.6],np.full(4,0.1 ))   #Heizspannung in V
epsilon = (U_heiz*I_heiz)/(I_kühl*U_kühl) #Berechung epsilon
epsilon_uncer=unp.std_devs(epsilon)
I_kühl_uncer=unp.std_devs(I_kühl)
epsilon_true=unp.nominal_values(epsilon)
I_kühl_true=unp.nominal_values(I_kühl)
print(epsilon)
plt.scatter(I_kühl_true,epsilon_true,color='black',marker='x') #mache Plot
plt.errorbar(I_kühl_true, epsilon_true, yerr = epsilon_uncer , xerr = I_kühl_uncer, fmt ='o')
plt.xlabel('Stromstärke Kühlung in A')
plt.ylabel('Leistungszahl')
plt.savefig('figures/Aufgabe2_2.png')
plt.show()

```

In dem Leistungszahl-$I_{kühl}$-Plot lässt sich erkennen, dass die Leistungsziffer bei zunehmendem Strom im Petierblock abnimmt. Dies liegt daran, dass die joulesche Wärme der Kühlung entgegenwirkt und die Kühlleistung bei höherem Stromfluss deutlich beeinträchtigt. Es ist sichtbar, dass die Kühlung effizienter bei geringerem Stromfluss ist.

Sobald Sie eine Kontaktstelle des Thermoelements mit einem Bunsenbrenner heizen und die andere mit Eiswasser kühlen können Sie mit der Konstruktion mit nur einer Leiterschleife durch ein Eisenjoch ein Gewicht von $m=5\,\mathrm{kg}$ tragen.

Vergleichen Sie die im Experiment beobachtete Tragkraft des Elektromagneten mit einem näherungsweise errechneten Wert. Berechnen Sie aus den geometrischen Abmessungen und dem spezifischen Widerstand von $\mathrm{Cu}$ den Widerstand der $\mathrm{Cu}$-Schleife. Damit ergibt sich der Thermostrom und der Betrag der [magnetischen Feldstärke](https://de.wikipedia.org/wiki/Magnetische_Feldst%C3%A4rke) $\vec{H}$ im Eisen. Schätzen Sie die mittlere Länge der Feldlinien im Eisen ab. Mit dem Erfahrungswert $\mu_{r}\approx500$ für die [relative Permeabilität](https://de.wikipedia.org/wiki/Magnetische_Permeabilit%C3%A4t) von Eisen für Magnete können Sie den Betrag der magnetischen Feldstärke $\vec{B}$ abschätzen. Eine begründete Formel für die Tragkraft findet sich in "Gerthsen, Kneser, Vogel: Probleme aus der Physik, Aufgabe 7.2.7". Dafür wird zusätzlich die Größe der Auflagefläche des Eisenjochs benötigt.

**Lösung:**
Bei dem letzten Versuch wird ein Magnetfeld in einer Leiterschleife durch einen Thermostrom
induziert. Dieses Magnetfeld soll ein Gewicht von 5 kg halten. Dafür wird eine Seite des Leiters
erwärmt, das andere Ende mit Eiswasser gekühlt wird. Das andere Ende wird mit Hilfe eines Bunsenbrenners erwärmt. Es entsteht eine induzierte Spannung. Dann wird der Bunsenbrenner entfernt. Es nimmt der induzierte Stromfluss ab, dies fürht zu einer kontinuierlichen Abschwächung des magnetischen Feld.
Wenn das Gewicht fällt ($F_{g}>F_{mag}$) wird die Spannung aufgenommen. Mit dieser Grenzspannung wird die Tragkraft berechnet. Die Formel lautet:
$$
F_{thermo}=\frac{A_{joch}\cdot \mu_{0} \cdot \mu_{r}^2 \cdot \left(U_{thermo}-U_0 \right)^2\cdot A_{schleife}^2}{8 \cdot r_{schleife}^2\cdot \rho_{Cu}^2\cdot L_{schleife}^2}
$$

```python
U_0 = ufloat(0.00018,0.00018*0.01) #messgerät offset
U_th  = ufloat(0.00145,0.00145*0.01 ) #spannung ab der es nicht mehr hält. 
rho_Cu = ufloat(1.7 * 10 **-8,0 ) #ohm/meter
r_joch = ufloat(0.035,0.035*0.01) #joch radius
A_joch = np.pi*r_joch**2

r_schleife = ufloat(0.065,0.065*0.01) # "radius" in meter der leiterschleife
L_schleife = 2* r_schleife# "länge" der leiterschleife
#mu_0 = 4* np.pi * 10**-7# -> alte Definition
mu_0 = ufloat(1.25663706212*10**(-6),0.0000000000019*10**(-6))  
mu_r = ufloat(500,0)
A_schleife = ufloat((8*10**-3)**2, ((8*10**-3)**2)*0.05) #querschnittsfläche der leiterschleife

F_thermo = A_joch * mu_0 * mu_r**2 * (U_th-U_0)**2 * A_schleife **2 /(8 * r_schleife**2 * rho_Cu**2 * L_schleife**2)

print(F_thermo , "N")
```

Aus unserer Berechnung ergibt sich eine Kraft von $48 \pm 5 \mathrm{N}$. 

Als theoretischer Wert ergibt sich $F = m\cdot g = 5\mathrm{kg} \cdot 9.81\mathrm{\frac{m}{s^2}} = 49.05\mathrm{N}$.\
Der berechnete Wert liegt im Konfidenzintervall. Daraus ergibt sich, dass die Fehlergrößen passend abgeschätzt wurden.
Der angegebene Mittelwert und Unsicherheit berücksichtigt aber nicht die Abweichung, welche durch die Näherung der Leiterschleife als Kreis entsteht. Somit ist der errechnete Wert nicht sehr aussagekräftig.
